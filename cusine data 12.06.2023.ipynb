```python
data.rename(columns={'Food Rating':'FoodRating','Alcohol ':'Alcohol', 'Service Rating':'ServiceRating', 'Overall Rating':'OverallRating','Often A S':'OftenAS'}, inplace=True)
data.head(10)
```

### Adding new column named age

```python
data['Age'] = [(2023 - YOB) for YOB in data['YOB']]
data.head()
```

```python
data.drop(columns=['Area code','User ID','YOB'], inplace=True)
data.head()
```

### (1) Persons from which location gives highest overall rating for Seafood cusines ?

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation = 90)
sns.barplot(data=data, y = 'OverallRating', x = 'Location', hue = 'Cuisines')
plt.show()
```

### (9)Within highest budget persons drinkers or non drinkers give highest rating ?

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.title('non drinkers')
sns.barplot(data=non_drinkers, x='Budget',y = 'OverallRating')

plt.subplot(1,2,2)
plt.title('often drinkers')
sns.barplot(data=often_drinkers, x='Budget',y = 'OverallRating')
plt.show()
```

##### Here from graph we can find that within highest budget persons often drinkers give highest rating than non alcoholic consumers

```python
sns.scatterplot(data=often_drinkers, y= 'Budget', x= 'OverallRating', hue='Cuisines')
plt.legend(loc='lower right')
plt.show()
```

##### Here from graph we can find that the persons having higher budget and giving good ratings are using French cuisines

```python
sns.barplot(data=often_drinkers,y= 'Budget', x= 'OverallRating', hue='Cuisines');
```

##### Here from this graph we could identify that the Filpino cuisines visitors give higher rating with highest budget.But, The filpino cuisines visitors give higher amount of weak rating. So, it's good to choose french cuisines to start a hotel business

### To find the percentage of cusines individually using pie chart

```python
cuisine_counts = data['Cuisines'].value_counts()

plt.figure(figsize=(8, 8))
sns.set_palette('pastel')
plt.pie(cuisine_counts, labels=cuisine_counts.index, autopct='%1.1f%%')
plt.title('Cuisine Distribution')
plt.show()
```
